```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline # Line allows for automatic showing of plots, otherwise need to use plt.imshow()
```

```python
import numpy as np
```

```python
x = np.linspace(0, 9, 10)
```

```python
x
```

```python
y = x ** 3
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'bo') # Similar to Matlab, b = blue, o = dots
plt.xlabel('X Label') # X Label
plt.ylabel('Y Label') # Y Label
plt.title('Title of the plot') # Title
plt.show() # Doesn't show the memory reference
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r+')
plt.title('Title 1')

plt.subplot(1,2,2)
plt.plot(y,x,'b--')
plt.title('Title 2')
plt.show() # Only one plt.show() is required at the end
```

```python
# Object oriented plotting
fig = plt.figure()
axes = fig.add_axes([0,0,1,1]) # Add axes left, bottom, width, height - All in fractions
axes.plot(x,y) # Doesn't show if in different line?
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title of the plot')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y)
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Title of the plot')

axes2 = fig.add_axes([0.1,0.4,0.5,0.5])
axes2.plot(y,x,'r--')
axes2.set_xlabel('X Label')
axes2.set_ylabel('Y Label')
axes2.set_title('Title of the plot')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3) #It's a numpy array (i.e. [][] for index)
axes[0][0].plot(x,y)
axes[1][1].plot(y,x,'r--')

fig
fig.tight_layout()
```

```python
fig, axes = plt.subplots(figsize=(12,4)) #Change the figure size
axes.plot(x,y)
```

```python
fig = plt.figure(figsize=(2,2),dpi=150) #For small figure size, increase DPI for better resolution P.S: Doesn't exist for subplot
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(figsize=(6,6))
axes.plot(y,x,'r--')
```

```python
fig.savefig('fig.png', dpi=300) #Figure is saved depending on extension but can pass the dpi argument as well
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, x**2, '^-', color='#00bf30', label='x**2', lw=0.5)
axes.plot(x, x**3, 'b--', label='x**3', linewidth=3, marker='.', markersize=20, markerfacecolor='green', markeredgecolor='red',markeredgewidth=2)
axes.legend(loc=7)
axes.set_xlim(2,7)
axes.set_ylim(10,300)
axes.grid()
```
